```python
import pandas as pd
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
file_path = "dataset_splits/ClassificationDataset-train0.xlsx"
df = pd.read_excel(file_path, names = ["labels", "text"])
data = df["text"].apply(lambda x: word_tokenize(x)).to_list()
ctr = 0
for i,sent in enumerate(data):
    for word in sent:
        if not f.emb(word.lower())[0]:
            
            print(word)
            print(f.emb(word.lower(), default="random"))
            ctr += 1
            break
            
#     print(i)
```
